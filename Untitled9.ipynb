```python
df["Postal Code"].value_counts()
```

```python
# Filter the DataFrame to include only postal codes with a value count of 1
unique_postal_codes = df['Postal Code'].value_counts()[df['Postal Code'].value_counts() == 1].index
filtered_df = df[df['Postal Code'].isin(unique_postal_codes)]

# Create the plot using Seaborn's relplot
sns.relplot(x="Postal Code", y="Profit", data=filtered_df, kind='scatter')
```

```python

plt.figure(figsize=(10, 6))
sns.barplot(x ='Sub-Category', y='Profit', data=df)
plt.title('Profit by Sub-Category')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (12, 6))
sns.barplot(x ='Region', y='Sales', data = df,estimator=sum)
plt.title('Sales by Region')
plt.show()
```

```python
plt.figure(figsize=(15,12))
df['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['green','blue'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

```python
ratioSales_Profit=df.groupby('Segment')['Profit'].sum()/df.groupby('Segment')['Sales'].sum()
```
