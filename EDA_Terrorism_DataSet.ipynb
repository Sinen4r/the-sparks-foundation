```python
# Group by 'attacktype1_txt' and count frequencies of 'targtype1_txt'
grouped = df.groupby('attacktype1_txt')['targtype1_txt'].value_counts().unstack(fill_value=0)

# Plotting each group
fig, axes = plt.subplots(nrows=len(grouped), figsize=(10, 20), sharex=True)
for i, (attack_type, frequencies) in enumerate(grouped.iterrows()):
    frequencies.plot(kind='bar', ax=axes[i], title=attack_type)
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
df.groupby('iyear')["nkill"].sum().plot(kind="bar")
plt.title("number of deaths each year")
plt.xlabel("year")
plt.ylabel("number of deaths")
```

```python
import plotly.express as px


cross_tab = pd.crosstab(df['iyear'], df['region_txt'])
plt.figure(figsize=(10, 6))
cross_tab = cross_tab.reset_index().melt(id_vars='iyear', var_name='region_txt', value_name='count')

# Plot using Plotly Express
fig = px.area(cross_tab, x='iyear', y='count', color='region_txt',
              title='terrorists activities each year',
              labels={'iyear': 'Year', 'count': 'Number of terrorists attack', 'region_txt': 'Region'},
              width=800, height=600)

fig.show()
```
